### Inner join
>* we will take out the common items from the two or more data set

```python
# student_id is common column in these two data set
students.merge(regs,how='inner',on='student_id')

# how parameter - which join we want to execute
# on parameter  - on which column we want to do joins 
# here students is our left table
# whereas regs is our right table 
```

### Outer join

```python
# it will display all the data from all the tables

students.merge(regs, how='outer', on='student_id')
```

### Left join

```python
courses.head()
```

```python
regs.head()
```

```python
courses.merge(regs, how='left', on='course_id')

# courses is left table and regs is right table
# here it will retain all the rows from left table
```

### Right join

```python
students
```

```python
temp_df = pd.DataFrame({
    'student_id':[26,27,28],
    'name':['Nitish','Ankit','Rahul'],
    'partner':[28,26,17]
})

students = pd.concat([students,temp_df],ignore_index=True)
students
```

```python
students.merge(regs, how='right', on='student_id')

# students is left table and regs is right table
# all the itmes from right table(regs) will be displayed
```

>* **find top 3 studiums with highest sixes/match ratio**

```python
matches
```

```python
delivery
```

```python
'''in the above two DataFrames we have common column named match_id'''

delivery.merge(matches, left_on='match_id', right_on='id')

# now we have data for each ball with stadium
```

```python
# we will save that DataFrame in variable temp_df

temp_df = delivery.merge(matches, left_on='match_id', right_on='id')
```

```python
temp_df[temp_df['batsman_runs'] == 6]

# we will get only those rows where runs scored is 6
```

```python
# we will save the above code in the variable six_df
six_df = temp_df[temp_df['batsman_runs'] == 6]

# now want sixes per stadium
# so now we can do groupby on the stadium column

six_df.groupby('venue')['venue'].count()
```

```python
# we can save this in the num_sixes variable
num_sixes = six_df.groupby('venue')['venue'].count()
```

```python
matches_per_venue = matches['venue'].value_counts()
matches_per_venue
```

```python
(num_sixes/matches_per_venue).sort_values(ascending=False).head(3)
```

>* **find orange cap holder of all the seasons**

```python
temp_df
```

```python
temp_df.groupby(['season','batsman'])
```

```python
temp_df.groupby(['season','batsman'])['batsman_runs'].sum()
```

```python
temp_df.groupby(['season','batsman'])['batsman_runs'].sum().reset_index()
```

```python
temp_df.groupby(['season','batsman'])['batsman_runs'].sum().reset_index(
).sort_values('batsman_runs',ascending=False)
```

```python
'''in the above DataFrame we have batsman highest score list 
overall, not season wise
to get that we will apply the drop_duplicates on season column
then in keep parameter we will keep the first one.
so eventually we will have the highest scorer batsman 
for each season and others will be dropped'''

temp_df.groupby(['season','batsman'])['batsman_runs'].sum().reset_index(
).sort_values('batsman_runs',ascending=False).drop_duplicates(
subset=['season'], keep='first').sort_values('season')
```
