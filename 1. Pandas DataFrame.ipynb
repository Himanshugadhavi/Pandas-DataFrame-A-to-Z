## `Selecting rows from a DataFrame`

- **iloc** - searches using index positions
- **loc** - searches using index labels 

### 1. iloc

#### single row

```python
movies.head(3)
```

```python
# suppose we want first row
movies.iloc[0]
# it will be a series because it's a single row
```

```python
type(movies.iloc[0])
```

#### multiple row

```python
movies.iloc[0:5]
# this will be dataframe data type
```

```python
# we want alternate movies between 5 to 15
movies.iloc[5:15:2]
```

#### fancy indexing 

```python
movies.iloc[[0,4,7,8]]
# we are passing list of items we want
```

### 2. loc
* when we provide range for fetching multiple items then the last number of given range will be included where as in iloc the last number is not included

```python
student_dict = {
    'name':['nitish','ankit','rupesh','rishabh','amit','himanshu'],
    'iq':[100,90,120,80,0,0],
    'marks':[80,70,100,50,0,0],
    'package':[10,7,14,2,0,0]
}

students = pd.DataFrame(student_dict)
students.set_index('name',inplace=True)
# we are setting name columns as our index
students
```

#### single row

```python
students.loc['rupesh']
```

#### multiple rows

```python
students.loc['nitish':'rishabh']
```

```python
students.loc['nitish':'rishabh':2]
# it will print alternare rows because step value is 2
```

#### fancy indexing 

```python
students.loc[['nitish','rupesh','himanshu']]
```

* **Note :** we can also use iloc on the students data though we have name column as index, because there is default index by pandas so here 0 and nitish will point out on same row positon

```python
students.iloc[[1,3,5]]
```

>* Action movies with rating higher than 7.5

```python
movies.head()
'''here in data if we look in to genres,
there is combination of genres in that column'''
```

```python
movies['genres']
```

```python
'''in the above column that data type is string,
now we want only action movies out of that,
if we directly apply split(|) on that Series then 
it will throw an error,
so we need to convert it in string object method'''

movies['genres'].str.split('|')
```

```python
movies['genres'].str.split('|').apply(lambda x:'Action' in x)
# here we are checking if action genre is present or not
```

```python
mask1 = movies['genres'].str.split('|').apply(lambda x:'Action' in x)

# now we want movies with rating > 7.5

mask2 = movies['imdb_rating'] > 7.5
```

```python
movies[mask1 & mask2].head()
```

```python
# using containg function

mask1 = movies['genres'].str.contains('Action')

'''now we want movies with rating > 7.5'''

mask2 = movies['imdb_rating'] > 7.5

movies[mask1 & mask2].head()
```

## `Important DataFrame Functions`

#### astype
* it changes the data type of the given column

```python
ipl.info()
# here memory occupied by ipl dataset is 148.6 kb
```

```python
ipl['ID'].astype('int32')
# we changed the data type of ID column to int32
```

```python
ipl['ID'] = ipl['ID'].astype('int32')
```

```python
ipl.info()
# now memory occcupied by ipl dataset is 144.9 kb
```

```python
# we will convert season datatype to category data type
# this will further reduce the occupied memory size

ipl['Season'] = ipl['Season'].astype('category')
```

```python
ipl.info()
# here memory occupied is 139.0 KB
```
